```python
#this is a list of graphs extracted from a given image
graph_set = extract_graph_components(image)
```

```python
def avg_coord(graph):
    nodes = graph.nodes()
    coords = np.array([nodes[i]['o'] for i in nodes])
    #o = (y_coord, x_coord)
    return np.average(coords[:, 1]), np.average(coords[:, 0])
```

```python
avg_graph_coords = [avg_coord(graph) for graph in graph_set]
```
